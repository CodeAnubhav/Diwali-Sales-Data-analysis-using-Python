```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as  sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv' ,encoding ='unicode_escape')
```

```python
df.drop(['Status' , 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status' : 'Marriage_status'})
```

```python
df.describe()
```

```python
df.info()
```

```python
ax = sns.countplot( x = 'Gender' , data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_gen = df.groupby(['Gender'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

ax = sns.barplot(x= 'Gender' , y= 'Amount' , data = sales_by_gen)
```

```python
df.columns
```

```python
ax = sns.countplot( x = 'Age Group' , data = df, hue='Gender' )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_gen = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.barplot(x= 'Age Group' , y= 'Amount' , data = sales_by_gen)
```

```python
sales_by_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x= 'State' , y= 'Orders' , data = sales_by_state)
```

```python
sales_by_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x= 'State' , y= 'Amount' , data = sales_by_state)
```

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_marst = df.groupby(['Marital_Status' , 'Gender'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x= 'Marital_Status' , y= 'Amount' , data = sales_by_marst, hue = 'Gender' )
```

```python
sns.set(rc={'figure.figsize':(25,5)})

ax = sns.countplot( x = 'Occupation' , data = df )

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_by_occ = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x= 'Occupation' , y= 'Amount' , data = sales_by_occ )
```

```python
sns.set(rc={'figure.figsize':(25,5)})

ax = sns.countplot( x = 'Product_Category' , data = df)

for bars in ax.containers:  
    ax.bar_label(bars)
```

```python
sales_by_productcat = df.groupby(['Product_Category'] , as_index = False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x = 'Product_Category' , y= 'Amount' , data = sales_by_productcat )
```

```python
sales_by_productID = df.groupby(['Product_ID'] , as_index = False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x = 'Product_ID' , y= 'Orders' , data = sales_by_productID )
```
